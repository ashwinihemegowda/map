```python
import pandas as pd
import numpy as np 
from pandas import Series, DataFrame
import seaborn as sb
from scipy import stats
```

```python
df = pd.read_csv('1000_movies_data.csv')
```

```python
# 1a. Retrieve only "Action" Movies
df[df['Genre']=='Action']
```

```python
# 1b. Retrieve only "Action + Sci-Fi" Movies
df.loc[df['Genre']=='Action,Sci-Fi']
#df.loc[(df['Genre'].str.contains('Action')) & (df['Genre'].str.contains('Sci-Fi')) ]
```

```python
# 1c. Retrieve the list of Movies which are released after the Year 2015
df[df['Year']>2015]
```

```python
# 1d. Retrieve the movies which has the maximum runtime (in Minutes)
df[df['Runtime (Minutes)']==df['Runtime (Minutes)'].max()]
```

```python
# 1e. Retrieve the list of movies which has got rating more than 7.5
df[df['Rating']>7.5]
```

```python
# 1f. Retrieve the top 3 movies which has got more revenue (in Millions) in 2014, 2015 and 2016
df[(df['Year']==2014)|(df['Year']==2015)|(df['Year']==2016)].sort_values(by='Revenue (Millions)',ascending=0).groupby('Year').head(3).sort_values(by='Year')
# 1f. Retrieve the top 3 movies which has got more revenue (in Millions) in 2014, 2015 and 2016
#df[df['Year'].isin([2014,2015,2016])].sort_values(by='Revenue (Millions)',ascending=0).groupby('Year').head(3).sort_values(by='Year')
```

```python
# 1g. Retrieve the top 3 movies in which "Will Smith" acted and has rating more than 7.0
df[(df['Actors'].str.contains('Will Smith')) & (df['Rating']>7.0)].sort_values(by='Rating' , ascending=0).head(3)
```

```python
# 1h. Retrieve the list of movies which is directed by "James Gary" and has metascore more than 90
df[(df['Director'].str.contains('James Gray')) & (df['Metascore']>70)]
```

```python
# 1i. Find the total numberof null entries in each and every columns
df.isnull().sum()
```

```python
# 1j. Retrieve the list of movies whose names are starting with "The"
df[df['Title'].str.startswith('The')]
```

```python
# 2a. Assign the dummy column names to each column (for eg. col_1, col_2....,col_9)
column_names=['col_1','col_2','col_3','col_4','col_5','col_6','col_7','col_8','target_label']
df2=pd.read_csv('diabetes.csv')
df2.columns=column_names
df2
```

```python
# 2b. Retrieve the count of each Target labeli.e, how many 0's and how many 1's
df2.groupby('target_label').size()
```

```python
# 2c. Analyse the outliers on every input column and visualize the outliers via "box plot"
sb.boxplot(x='col_1' ,data=df2)
#Finding Z Score => (data_point-mean)/standard_deviation
mean_col_1=np.mean(df2['col_1'])
print('mean_col_1 -',mean_col_1)
std_col_1 = np.std(df2['col_1'])
print('std_col_1 -',std_col_1)
z_score_col_1 = np.abs(stats.zscore(df2['col_1']))
print('Outliers for col_1 -',np.where(z_score_col_1 >3))
print('-'*60)

#Finding IQR =>Quartile3 - Quartile1
Q1_col_1 = np.percentile(df2['col_1'],25,method='midpoint')
Q3_col_1 = np.percentile(df2['col_1'],75,method='midpoint')
IQR_col_1 = Q3_col_1 - Q1_col_1
print('IQR_col_1 -',IQR_col_1)
#Upper and Lower bound
upper_col_1 = Q3_col_1 + 1.5 * IQR_col_1
lower_col_1 = Q1_col_1 - 1.5 * IQR_col_1
upper_array = np.where(df2['col_1'] >= upper_col_1)
lower_array = np.where(df2['col_1'] <= lower_col_1)
print('Outliers in Upper Array -',upper_array)
print('Outliers in Lower Array -',lower_array)
```
